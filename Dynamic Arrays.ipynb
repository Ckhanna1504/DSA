```python
"""
Dynamic Array

1. Create List [D]
2. len[D]
3. append[d]
4. print[d]
5. indexing[d]
6. pop
7. clear
8. find
9. insert
10. delete
11. remove

sort/ min/ max/ sum
extend
negative indexing

"""
```

```python
import ctypes
```

```python
class myList:
    def __init__(self):
        self.size = 1 #how many elements can be stored in the array
        self.n = 0 # how many elements are currently stored in the array

        self.A = self._create_array(self.size)
    
    def __len__(self):
        return self.n
    
    def append(self, item):
        if self.n == self.size:
            
            self.__resize(self.size*2)
            
        self.A[self.n] = item
        self.n = self.n + 1

    def __resize(self, newcapacity):
        
        # creating temp array with new size
        B = self._create_array(newcapacity)
        self.size = newcapacity
        # copying elements
        for i in range(self.n):
            B[i] = self.A[i]
        
        self.A = B
        
    def __str__(self):
        
        if self.n==0:
            return "[]"
        
        print_str = ""
        for i in range(self.n):
            print_str = print_str + str(self.A[i]) + ","
        
        return "[" + print_str[:-1] + "]"
    
    def __getitem__(self, key):
        if key < self.n:
            return self.A[key]
        else:
            return "Index Error: Index out of bound"
    
    def pop(self, pos = None):
        
        if pos is None:
            pos = self.n-1

        if self.n==0 or pos>=self.n:
            return "Index Error: No elements to pop."
        
        pop_value = self.A[pos]
        B = self._create_array(self.size)
        
        i=0
        j=0

        while i < self.n:
            if i==pos:
                i=i+1
                B[j] = self.A[i]
            else:                
                B[j] = self.A[i]
                i=i+1
                j=j+1
        
        self.A = B
        self.n = self.n - 1
        return pop_value

    def clear(self):
        self.size=1
        self.n=0
        

    def find(self, item):
        for i in range(self.n):
            if item == self.A[i]:
                return i
            
        return "Not found Error: Value not found!"
    
    def insert(self, index, item):
        
        if self.n == self.size:
            self.__resize(self.size*2)
        
        for i in range(self.n, index, -1):
            self.A[i] = self.A[i-1]
        
        self.A[index] = item
        self.n = self.n + 1

    def __delitem__(self, pos):
        if pos >= self.n:
            return "Index Error: Index out of bound."
        
        for i in range(pos, self.n-1):
            self.A[i] = self.A[i+1]
            
        self.n = self.n - 1 
        
    def remove(self, item):
        pos = self.find(item)
        if isinstance(pos, int):
            self.__delitem__(pos)
        else:
            return pos

    def sort(self):
        pass        

    def min(self):
        
        if self.n==0:
            return "Value Error: No items to evaluate."
        
        min = self.A[0]
        for i in range(self.n):
            if isinstance(self.A[i], str) or isinstance(self.A[i], bool):
                return "Type Error: Cannot compute string."
            if min>self.A[i]:
                min=self.A[i]
        return min

    def max(self):
        max = 0
        if self.n==0:
            return "Value Error: No items to evaluate."
        for i in range(self.n):
            if isinstance(self.A[i], str) or isinstance(self.A[i], bool):
                return "Type Error: Cannot compute string."
            if max<self.A[i]:
                max=self.A[i]
        return max

    def sum(self):
        
        if self.n==0:
            return "Value Error: No items found."
        
        sum=0
        for i in range(self.n):
            
            if isinstance(self.A[i], str) or isinstance(self.A[i], bool):
                continue
            
            sum = sum + self.A[i]
        return sum
    
    def __iter__(self):
        if self.n==0:
            return "Value Error: No items found."
        for i in range(self.n):
            if isinstance(self.A[i], str) or isinstance(self.A[i], bool):
                continue
            yield self.A[i]
        
        

    def extend(self):
        pass          
    
    def negative_indexing(self):
        pass
    
    def slicing(self):
        pass

    def merge(self):
        pass

    def _create_array(self, size):
        return (size*ctypes.py_object)()
    
```
