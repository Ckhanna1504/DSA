```python
"""
Linked List:
Node[D] - done
Create a LL - done
len
insert from Head - done
traverse/print - done
insert from tail (append) - done
insert in middle(after) - done

clear - done
delete from head - done
delete from tail(pop) - done
delete by value(remove) - done

search by value(find) - done 
delete by index -> del l[0] - done 
search by index(indexing)

"""
```

```python
from itertools import count


class Node:

    def __init__(self, value):
        self.data = value
        self.next = None
        
class LinkedList:
    def __init__(self):
        # empty Linked List
        self.Head = None
        self.Tail = None
        # number of elements in the linked list
        self.n=0

    def __len__(self):
        return self.n
    
    def insert(self, item):
        if self.n==0:
            newNode = Node(item)
            self.Head = newNode
            self.Tail = newNode

        else:
            new_node = Node(item)
            new_node.next = self.Head
            self.Head = new_node
        self.n = self.n + 1
    
    def append(self, item):
        
        if self.n==0:
            newNode = Node(item)
            self.Head = newNode
            self.Tail = newNode

        elif self.n == 1:
            self.Tail = Node(item)
            self.Head.next = self.Tail
            
        else:
            new_node = Node(item)
            self.Tail.next = new_node
            self.Tail=new_node

        self.n = self.n + 1

    def after(self, idx, item):
        if idx > self.n:
            return "Index Error: Index out of bound."
        
        elif idx==0:
            self.insert(item)
        
        elif idx == self.n - 1:
            self.append(item)

        else:
                        
            count = 0
            currNode = self.Head
            while count < idx:
                currNode = currNode.next
                count = count + 1

            newNode = Node(item)
            newNode.next = currNode.next
            currNode.next = newNode

    def clear(self):
        self.Head = None
        self.n= 0

    def delete(self): # delete from head
        if self.n == 0 :
            return "Index Error: No element to delete"
        elif self.n == 1:
            val = self.Head.data
            self.Head = None
        else:
            val = self.Head.data
            self.Head = self.Head.next
        self.n = self.n - 1
        return val

    def pop(self):
        if self.n == 0:
            return "Index Error: No element to delete"
        
        elif self.n == 1 :
            val = self.Head.data
            self.Head = None
            self.Tail = None
        
        else:
            
            currNode = self.Head

            while currNode.next != self.Tail:
                currNode = currNode.next
            
            val = self.Tail.data
            
            self.Tail = currNode
            self.Tail.next = None
        self.n = self.n - 1
        return val

    def remove(self, item):
        if self.Head.data == item:
            self.Head = self.Head.next
            self.n = self.n - 1
            return
        
        currNode = self.Head
        while currNode.next != None:

            if currNode.next.data == item:
                if currNode.next == self.Tail:
                    currNode.next = None
                    self.Tail = currNode
                else:
                    currNode.next = currNode.next.next
                self.n = self.n - 1
                return
            
            currNode = currNode.next
                    
        return "Value Error: Value not found."

            
    def find(self, item):
                
        currNode = self.Head
        count = 0
        while currNode != None:

            if currNode.data == item:
                return count
            
            currNode = currNode.next
            count = count + 1
                    
        return "Value Error: Value not found."

    def __delitem__(self, key):
        if int(key) >= self.n:
            print("Index Error: Index out of bound")
        
        elif key == 0:
            
            val = self.Head.data
            self.Head = self.Head.next
            self.n = self.n - 1
            print(val)
        
        count = 1
        currNode = self.Head

        while currNode != None:
            if key == count:
                if count == self.n - 1:
                    val = currNode.next.data
                    currNode.next=None
                    self.Tail = currNode
                else:
                    currNode.next = currNode.next.next
                self.n = self.n - 1
                print(val)
            currNode = currNode.next
        
        
    def __str__(self):
        
        if self.n==0:
            return "[]"
        
        else:
            
            curr_node = self.Head
            str_print = ""
            
            while curr_node != None:
                
                str_print = str_print + str(curr_node.data) + ","
                
                curr_node = curr_node.next

            return "[" + str_print[:-1] + "]"
        
    def __getitem__(self, key):
        
        
        start = key.start
        stop = key.stop
        step = key.step
        
        if isinstance(key, int):
            stop = None
            step = 1
        elif isinstance(key, slice):
            
            if key.start is None:
                start = 0
                
            
            if key.stop is None:
                stop = self.n
                
            if key.step is None:
                step = 1
                
        if stop > self.n or start > self.n:
            raise IndexError("Index out of bound")
        
        i=0
        indexStr = ""
        currNode = self.Head

        while i < self.n:
            
            print(i)

            if isinstance(key, int) and i == key:
                return currNode.data
                
            if isinstance(key, slice):
                if i>=start and i<=stop:
                    indexStr = indexStr + str(currNode.data) + ","
                    i = i + step
                    
                    for _ in range(step):
                        if currNode.next is None :
                            break
                        currNode = currNode.next
                else:
                    i += 1
                    currNode = currNode.next
           
        return "[" + indexStr[:-1] + "]"               
```
